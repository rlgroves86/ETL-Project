```python
#Dependencies

from splinter import Browser
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import pandas as pd
import requests
import os
```

```python
#Load in the website

executable_path = {'executable_path': '/Applications/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://gregslist.com/phoenix/software-companies-other/woman-ceo/'
browser.visit(url)
```

```python
html = browser.html
soup = BeautifulSoup(html, 'lxml')
#companies = soup.find_all('div', class_='list_text')
```

```python
#Print the code

print(soup.prettify())
```

```python
#Show the coding for the first company on the list

company_listings = soup.find_all('div', class_='company-listing')

print(company_listings[0])
```

```python
#Show the names of the companies

title = [x.text.strip() for x in soup.find_all('div', class_='col-md-3 company-logo-column')]


title
```

```python
#Show the details of each company

detail = [xe.text.strip()for xe in soup.find_all ('span')]

detail
```

```python
#List all of the companies and details

software = []

for company in company_listings:
    title = company.find('a').text.strip()
    website = company.find('a')['href']
    detail = company.find('span', class_='detail')
    if detail: 
        detail = detail.text.strip()
    else:
        detail = 'None'
    
   
    software.append({
        'title': title,
        'detail': detail,
        'website': website
})
     

software_df = pd.DataFrame(software)

software_df = software_df.rename(columns={"title": "Company",
                                    "detail": "Detail",
                                    "website": "Website"})

software_df
```

```python
#Import the information to a .csv

software_df.to_csv("women_ceo.csv", index=False, header=True)
```
